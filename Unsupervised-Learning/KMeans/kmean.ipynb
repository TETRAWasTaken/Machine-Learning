# K Means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.metrics import (silhouette_score, normalized_mutual_info_score,
                                davies_bouldin_score, silhouette_samples)
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.spatial.distance import cdist
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

df=pd.DataFrame(X, columns=data.feature_names)
display(df)
```

## Applying StandardScalar

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

## Elbow Curve

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid()
plt.show()
```

## Training a Single Model

```python
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
display(df.head())

centroids = kmeans.cluster_centers_
pcs = PCA(n_components=2)
X_pca = pcs.fit_transform(X_scaled)

centroids_pca = pcs.transform(centroids)

plt.Figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='k', s=100)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (PCA Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()
plt.show()
```

## Within Cluster Sum of Squares ( WCSS )

```python
wcss = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

wcss_score = (wcss - np.min(wcss)) / (np.max(wcss) - np.min(wcss))
print(wcss_score)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss_score, marker='o')
plt.title('WCSS Normalised Score')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid()
plt.show()
```

## Evaluation Metrics

```python
sil_score = silhouette_score(X_scaled, clusters)
print(f'Silhouette Score: {sil_score:.4f}')

db_score = davies_bouldin_score(X_scaled, clusters)
print(f'Davies-Bouldin Score: {db_score:.4f}')
```

## Dunn Index

```python
def dunn_index(X, labels):
    clusters = np.unique(labels)
    intra = []
    inter = []

    for c in clusters:
        cluster_points = X[labels == c]
        intra.append(np.max(cdist(cluster_points, cluster_points)))

    for i in range(len(clusters)):
        for j in range(i + 1, len(clusters)):
            cluster_i = X[labels == clusters[i]]
            cluster_j = X[labels == clusters[j]]
            inter.append(np.min(cdist(cluster_i, cluster_j)))

    return np.min(inter) / np.max(intra)


dunn_score = dunn_index(X_scaled, clusters)
print(f'Dunn Index: {dunn_score:.4f}')
```
