# t-SNE dimensionality reduction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
```

```python
data = pd.read_csv("/workspaces/Supervised-Machine-Learning/Datasets/breast+cancer+wisconsin+diagnostic/wdbc.data", header=None)
display.display(data)
```

## Imports

```python
import pylab as pl
from sklearn import manifold, datasets, preprocessing
from functools import partial 
import time
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from IPython import display
```

```python
data.iloc[:, 1] = data.iloc[:, 1].map({"M": 1, "B": 0})
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values
```

```python
display.display(X)
display.display(y)
```

```python
for i in range(250, 500, 2):
    fig = plt.figure(figsize=(10, 10))

    xe = manifold.TSNE(n_components=2, max_iter=i, perplexity=30, random_state=0).fit_transform(X)

    plt.scatter(xe[:, 0], xe[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), s=50)
    plt.colorbar(ticks=range(10))

    for i, txt in enumerate(y):
        if i % 20 == 0:
            plt.annotate(txt, (xe[i, 0], xe[i, 1]))
    
    plt.axis("off")
    display.clear_output(wait=True)
    display.display(pl.gcf())

    plt.clf()
    plt.cla()
    plt.close('all')
    
```

```python
for i in range(250, 500, 2):
    fig = plt.figure(figsize=(10, 10))

    xe = manifold.MDS(n_components=2, max_iter=50, n_init=4, random_state=0, normalized_stress=False).fit_transform(X)

    plt.scatter(xe[:, 0], xe[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), s=50)
    plt.colorbar(ticks=range(10))

    for i, txt in enumerate(y.tolist()):
        if i % 20 == 0:
            plt.annotate(txt, (xe[i, 0], xe[i, 1]))
    
    plt.axis("off")
    display.clear_output(wait=True)
    display.display(pl.gcf())

    plt.clf()
    plt.cla()
    plt.close('all')
    
```

```python
perplexity_low = 5
perplexity_high = 50

tsne_low = manifold.TSNE(n_components=2, perplexity=perplexity_low, random_state=0).fit_transform(X)
embedding_low = tsne_low.fit_transform(X)

tsne_high = manifold.TSNE(n_components=2, perplexity=perplexity_high, random_state=0).fit_transform(X)
embedding_high = tsne_high.fit_transform(X)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(embedding_low[:, 0], embedding_low[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), s=50, label='0')
plt.scatter(embedding_low[:, 0], embedding_low[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), s=50, label='1')
plt.title(f"t-SNE with Perplexity={perplexity_low}")
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(embedding_high[:, 0], embedding_high[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), s=50, label='0')
plt.scatter(embedding_high[:, 0], embedding_high[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), s=50, label='1')
plt.title(f"t-SNE with Perplexity={perplexity_high}")
plt.legend()
```
