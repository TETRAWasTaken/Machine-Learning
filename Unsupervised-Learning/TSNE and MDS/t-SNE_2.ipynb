# t-SNE dimensionality reduction on Iris dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
```

```python
data = pd.read_csv("/workspaces/Supervised-Machine-Learning/Datasets/Iris.csv")
display.display(data.head())
```

## Imports

```python
import pylab as pl
from sklearn import manifold, preprocessing
from sklearn.decomposition import PCA
import time
from IPython import display
```

```python
label_encoder = preprocessing.LabelEncoder()
y = label_encoder.fit_transform(data["Species"] )
X = data[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
```

```python
display.display(X)
display.display(y)
```

```python
for i in range(250, 400, 10):
    fig = plt.figure(figsize=(10, 10))

    xe = manifold.TSNE(n_components=2, max_iter=i, perplexity=30, random_state=0).fit_transform(X)

    plt.scatter(xe[:, 0], xe[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 3), s=50)
    plt.colorbar(ticks=range(3))

    for j, txt in enumerate(y):
        if j % 15 == 0:
            plt.annotate(txt, (xe[j, 0], xe[j, 1]))

    plt.axis("off")
    display.clear_output(wait=True)
    display.display(pl.gcf())

    plt.clf()
    plt.cla()
    plt.close('all')
```

```python
for i in range(250, 400, 10):
    fig = plt.figure(figsize=(10, 10))

    xe = manifold.MDS(n_components=2, max_iter=50, n_init=4, random_state=0, normalized_stress=False).fit_transform(X)

    plt.scatter(xe[:, 0], xe[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 3), s=50)
    plt.colorbar(ticks=range(3))

    for j, txt in enumerate(y):
        if j % 15 == 0:
            plt.annotate(txt, (xe[j, 0], xe[j, 1]))

    plt.axis("off")
    display.clear_output(wait=True)
    display.display(pl.gcf())

    plt.clf()
    plt.cla()
    plt.close('all')
```

```python
perplexity_low = 5
perplexity_high = 40

tsne_low = manifold.TSNE(n_components=2, perplexity=perplexity_low, random_state=0)
embedding_low = tsne_low.fit_transform(X)

tsne_high = manifold.TSNE(n_components=2, perplexity=perplexity_high, random_state=0)
embedding_high = tsne_high.fit_transform(X)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for ax, embedding, perplexity in [
    (axes[0], embedding_low, perplexity_low),
    (axes[1], embedding_high, perplexity_high),
]:
    for label, color in [(0, "tab:blue"), (1, "tab:orange"), (2, "tab:green")]:
        mask = y == label
        ax.scatter(
            embedding[mask, 0],
            embedding[mask, 1],
            s=40,
            alpha=0.8,
            c=color,
            label=f"Class {label}"
        )
    ax.set_title(f"t-SNE with Perplexity={perplexity}")
    ax.set_xlabel("Component 1")
    ax.set_ylabel("Component 2")
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
results = pd.DataFrame({
    "Perplexity": [perplexity_low, perplexity_high],
    "KL Divergence": [tsne_low.kl_divergence_, tsne_high.kl_divergence_]
})

display.display(results)

best_idx = results["KL Divergence"].idxmin()
best_perplexity = results.loc[best_idx, "Perplexity"]

print("Lower KL divergence indicates a better local neighborhood fit.")
print(f"Best perplexity in this run: {best_perplexity}")
```
