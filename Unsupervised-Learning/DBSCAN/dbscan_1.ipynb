## Applying Scaling

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
display(scaled_df)
```

## Applying DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.01, min_samples=5)
labels = dbscan.fit_predict(scaled_df)


df["Cluster"] = labels
display(df)
```

## Performance Metrics 

### Dunn Index

```python
def dunn_index(X, labels):
    unique_labels = np.unique(labels)
    clusters = [X[labels == label] for label in unique_labels if label != -1]

    if len(clusters) < 2:
        return 0.0

    intra_distances = []
    for cluster in clusters:
        # A cluster with 0/1 samples has zero diameter by definition.
        if len(cluster) < 2:
            intra_distances.append(0.0)
            continue
        pairwise_distances = np.linalg.norm(cluster[:, np.newaxis] - cluster, axis=2)
        intra_distances.append(np.max(pairwise_distances))

    inter_distances = []
    for i, cluster1 in enumerate(clusters):
        for j, cluster2 in enumerate(clusters):
            if i < j:
                pairwise_distances = np.linalg.norm(cluster1[:, np.newaxis] - cluster2, axis=2)
                inter_distances.append(np.min(pairwise_distances))

    if not inter_distances:
        return 0.0

    min_inter = np.min(inter_distances)
    max_intra = np.max(intra_distances)

    if max_intra == 0:
        return np.inf if min_inter > 0 else 0.0

    return min_inter / max_intra

dunn_score = dunn_index(scaled_df.values, labels)
print(f"Dunn Index: {dunn_score}")
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(scaled_df, labels)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(scaled_df, labels)
print(f"Davies-Bouldin Score: {db_score}")
```

## Applying GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

param_grid = {
    'eps': [0.01, 0.05, 0.1, 0.5, 1.0],
    'min_samples': [3, 5, 10]
}

def dbscan_silhouette_scorer(estimator, X, y=None):
    labels = estimator.fit_predict(X)

    unique_labels = set(labels)
    # Invalid for silhouette: only one label, all noise, or one label + noise.
    if len(unique_labels) <= 1 or unique_labels == {-1} or (len(unique_labels) == 2 and -1 in unique_labels):
        return -1.0

    try:
        return silhouette_score(X, labels)
    except ValueError:
        return -1.0

dbscan = DBSCAN()
grid_search = GridSearchCV(
    estimator=dbscan,
    param_grid=param_grid,
    scoring=dbscan_silhouette_scorer,
    cv=3
 )
grid_search.fit(scaled_df)

print("Best params:", grid_search.best_params_)
print("Best silhouette:", grid_search.best_score_)
```
