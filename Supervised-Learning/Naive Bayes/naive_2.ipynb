# Naive Bayes

```python
import pandas as pd
```

## Importing the Dataset

```python
dataset = pd.read_csv("/workspaces/Supervised-Machine-Learning/Datasets/adult.csv")
dataset.info()
```

```python
display(dataset)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Applying Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X = dataset.iloc[:, :-1].copy()
y = dataset.iloc[:, -1].copy()

categorical_features = [1, 3, 5, 6, 7, 8, 9, 13]
for feature in categorical_features:
    X.iloc[:, feature] = le.fit_transform(X.iloc[:, feature])

y_encoded = le.fit_transform(y)
X_encoded = X.to_numpy(dtype=float)
```

```python
from sklearn.preprocessing import RobustScaler
sc = RobustScaler()

numerical_features = [0, 2, 4, 10, 11, 12]
X_gaussian = X_encoded.copy()

for feature in numerical_features:
    X_gaussian[:, feature] = sc.fit_transform(X_gaussian[:, feature].reshape(-1, 1)).reshape(-1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_gaussian, y_encoded, test_size=0.25, random_state=324
)
```

## Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

## Multinomial Naive Bayes

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.25, random_state=324
)
```

## Categorical Naive Bayes

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.25, random_state=324
)
```

## Naive Bayes Score Output

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler, MinMaxScaler, Binarizer, KBinsDiscretizer
from sklearn.naive_bayes import GaussianNB, MultinomialNB, ComplementNB, BernoulliNB, CategoricalNB

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.25, random_state=324
)

scores = {}

# Gaussian Naive Bayes
rb = RobustScaler()
X_train_rb = X_train.copy()
X_test_rb = X_test.copy()
numerical_features = [0, 2, 4, 10, 11, 12]
for feature in numerical_features:
    X_train_rb[:, feature] = rb.fit_transform(X_train_rb[:, feature].reshape(-1, 1)).reshape(-1)
    X_test_rb[:, feature] = rb.transform(X_test_rb[:, feature].reshape(-1, 1)).reshape(-1)
model = GaussianNB()
model.fit(X_train_rb, y_train)
scores['GaussianNB'] = model.score(X_test_rb, y_test)

# Multinomial Naive Bayes
mm = MinMaxScaler()
X_train_mm = mm.fit_transform(X_train)
X_test_mm = mm.transform(X_test)
model = MultinomialNB()
model.fit(X_train_mm, y_train)
scores['MultinomialNB'] = model.score(X_test_mm, y_test)

# Complement Naive Bayes
model = ComplementNB()
model.fit(X_train_mm, y_train)
scores['ComplementNB'] = model.score(X_test_mm, y_test)

# Bernoulli Naive Bayes
binz = Binarizer(threshold=0.5)
X_train_bin = binz.fit_transform(X_train_mm)
X_test_bin = binz.transform(X_test_mm)
model = BernoulliNB()
model.fit(X_train_bin, y_train)
scores['BernoulliNB'] = model.score(X_test_bin, y_test)

# Categorical Naive Bayes
kbd = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')
X_train_cat = kbd.fit_transform(X_train)
X_test_cat = kbd.transform(X_test)
model = CategoricalNB()
model.fit(X_train_cat, y_train)
scores['CategoricalNB'] = model.score(X_test_cat, y_test)

pd.DataFrame(list(scores.items()), columns=['Model', 'Score']).sort_values(by='Score', ascending=False).reset_index(drop=True)
```
