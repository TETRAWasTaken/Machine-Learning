## Bias
Difference between the actual and predicted value(residual) of training dataset
* high bias
    * Not capturing training data >> Low Training Accuracy
* Low bias
    * Able to cature the training data >> High Training Accuracy

## Variance
Difference between two different dataset accuracy ( Testing and Training )
* High Variance
    * Low Testing Accuracy and High Training Accuracy
    * Low Training Accuracy and High Testing Accuracy
* Low Variance
    * High Training and Testing Accuracy
    * Low Training and Testing Accuracy

## Overfitting
Model will perform only on Training data, but perform poorly over Testing data

Overfitting occurs when the model has **Low Bias** and **High Variance**

## Underfitting
Model will perform poorly on both Training and Testing data

Underfitting occurs when the model has **High Bias** and **Low Variance**

## Avoiding Overfitting

**Hyper parameter tuning :-**
* GridSearchCV

**Early Stopping**

**Regularization :-**

* **Hyperparameter Tuning**

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function (e.g., f(x) = x^2) and its gradient
def func(x):
    return x ** 2


def gradient(x):
    return 2 * x


# Gradient Descent Parameters
learning_rate = 0.1
start_x = 4.0
iterations = 20

# Store history for plotting
x_history = [start_x]
y_history = [func(start_x)]

# Perform Gradient Descent
x = start_x
for _ in range(iterations):
    grad = gradient(x)
    x = x - learning_rate * grad
    x_history.append(x)
    y_history.append(func(x))

# Plotting
x_vals = np.linspace(-5, 5, 100)
y_vals = func(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Loss Function: $f(x) = x^2$')
plt.scatter(x_history, y_history, color='red', zorder=5, label='Gradient Descent Steps')
plt.plot(x_history, y_history, color='red', linestyle='--', alpha=0.5)

# Annotate start and end
plt.annotate('Start', xy=(x_history[0], y_history[0]), xytext=(x_history[0] + 0.5, y_history[0]),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Minimum', xy=(0, 0), xytext=(0, 5),
             arrowprops=dict(facecolor='green', shrink=0.05))

plt.title('Gradient Descent Visualization')
plt.xlabel('x (Parameter)')
plt.ylabel('f(x) (Loss)')
plt.legend()
plt.grid(True)
plt.show()
```
